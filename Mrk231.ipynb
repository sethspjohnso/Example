```python
%matplotlib inline
import sys, logging, os, re
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
from readcol import readcol
from scipy import stats
import scipy.constants as const
from satmc import satmc
```

```python
### Observed Mrk231 SED
l_Kis,f_Kis,ferr_Kis          = readcol("Mrk231_Kishimoto.dat",twod=False)
l_MMT,f_MMT,ferr_MMT          = readcol("Mrk231_MMT.dat",twod=False)
l_Ima,f_Ima,caca,ferr_Ima     = readcol("Mrk231_Imanishi.dat",twod=False)
l_CCSpec,f_CCSpec,ferr_CCSpec = readcol("Mrk231_CCSpec.dat",twod=False)
l_CCIma,f_CCIma,ferr_CCIma    = readcol("Mrk231_CCIma.dat",twod=False)

f_CCSpec = f_CCSpec*1000.
ferr_CCSpec = ferr_CCSpec*1000.
```

```python
###############
### Figures ###
###############
font_size=15.0
label_size=10.
lw = 3.
fig,(ax1) = plt.subplots(1,1,figsize=(20,10),dpi=120)

#Model
ax1.errorbar(l_Kis,f_Kis,yerr=ferr_Kis,fmt='o',color='red',markersize=10.,label='Kishimoto+09')
ax1.errorbar(l_MMT,f_MMT,yerr=ferr_MMT,fmt='s',color='black',markersize=10.,label='MMT-Pol: This work')
ax1.errorbar(l_Ima,f_Ima,yerr=ferr_Ima,fmt='o',color='grey',markersize=5.,label='Imanishi+10')
ax1.errorbar(l_CCSpec,f_CCSpec,yerr=ferr_CCSpec,fmt='o',color='blue',markersize=5.,label='Alonso-Herrero+16')
ax1.errorbar(l_CCIma,f_CCIma,yerr=ferr_CCIma,fmt='s',color='green',markersize=10.,label='CC-Pol: This work')

ax1.set_xlabel('Wavelength ($\mu$m)',fontsize=font_size)
ax1.set_ylabel('Flux (mJy)',fontsize=font_size)
ax1.set_ylim([10,4000])
ax1.set_xlim([1,13])
ax1.xaxis.set_tick_params(labelsize=font_size)
ax1.yaxis.set_tick_params(labelsize=font_size)
ax1.set_yscale('log')
#ax1.set_xscale('log')
ax1.legend(loc='best')
```

```python
###Concatenate arrays
l_mrk231 = np.concatenate((l_Kis,l_MMT,l_Ima,l_CCSpec,l_CCIma))
f_mrk231 = np.concatenate((f_Kis,f_MMT,f_Ima,f_CCSpec,f_CCIma))
ferr_mrk231 = np.concatenate((ferr_Kis,ferr_MMT,ferr_Ima,ferr_CCSpec,ferr_CCIma))
```

```python
### 2-phase Clumpy SED from  Siebenmorgen , Heymann, & Efstathiou A&A 2015,
ClumpySED_path = '/Users/enrique/Desktop/Clumpy_2phase_SHE15/Clumpy_2phase/agnsed/'
ClumpySED_list = ClumpySED_path+'ClumpySED.list' 
ClumpySED_list = readcol(ClumpySED_list)

nSEDm = len(ClumpySED_list)
nSEDpt = len(l_mrk231)

templates = np.empty(nSEDm,dtype=[('modelname','a29'),('Rin','f8'),('Vc','f8'),('Tau_c','f8'),
                                  ('Tau_m','f8'),('Angle','f8'),('lum','f8'),
                                   ('wavesim','f8',(nSEDpt,)),
                                   ('fluxsim','f8',(nSEDpt,)),
                                   ])
```

```python
###Fill the variables in the wrapper 
sed_zero = zeros(len(l_mrk231))

#Bandwidth to estimate the flux from the SED library of Siebenmorgen. 
#for now I just assume a bandwith of 0.3 um per photometric data. Just to check the code
Dl = 0.3

c    = 2.9979e14          #speed of light (microns/s)
dist = 174.6              #Mpc
z    = 70.*dist/3.e5      #Redshift

for ii in range(0,len(ClumpySED_list)):
    l_i,f_i = readcol(ClumpySED_path+ClumpySED_list[ii][0],twod=False)
    for jj in range(0,len(l_mrk231)):
        sed_zero[jj] = np.mean(f_i[np.where((l_i < l_mrk231[jj]+Dl) & (l_i > l_mrk231[jj]-Dl))])
        
    #Create array of templates
    string=re.findall(r'[-+]?\d*\.\d+|\d+',ClumpySED_list[ii][0])
    Rin   = string[0]
    Vc    = string[1]
    Tau_c = string[2]
    Tau_m = string[3]
    Angle = string[4]
    
    #Calculate fluxes & luminosities
    lnu = 4*3.1415*(dist/3.086E22)**2*sed_zero/(1.+z)              #L_sun
    lnu = sed_zero*(dist/3.086E22)**2./(1.+z)/(1.13e-8)/c
    lum = abs(np.trapz(lnu,c/l_mrk231))                              
    flux = sed_zero*(1.+z)*(lum)/(dist/3.086E22)**2.*1000          #mJy
    
    #Filling Templates
    templates[ii]['modelname'] = ClumpySED_list[ii][0]
    templates[ii]['Rin']       = Rin
    templates[ii]['Vc']        = Vc
    templates[ii]['Tau_c']     = Tau_c
    templates[ii]['Tau_m']     = Tau_m
    templates[ii]['Angle']     = Angle
    templates[ii]['lum']       = lum
    templates[ii]['wavesim'][:] = l_mrk231
    templates[ii]['fluxsim'][:] = flux
```

```python
satmc('Mrk231_SED.dat',templates,redshift=0.042)
```

```python
#Random choice of SED library combined with a BB function for future purposes of the SED fitting.

l_i,f_i = readcol(ClumpySED_path+ClumpySED_list[3409][0],twod=False)

print ClumpySED_list[3409][0]

def blackbody(lam, T):
    """ Blackbody as a function of wavelength (um) and temperature (K).

    returns units of erg/s/cm^2/cm/Steradian
    """
    from scipy.constants import h,k,c
    lam = 1e-6 * lam # convert to metres
    return 2*h*c**2 / (lam**5 * (np.exp(h*c / (lam*k*T)) - 1))

Tbb_AGN = 1200.
AGN=blackbody(l_i,Tbb_AGN)/7E7


### TOTAL SED
TOTAL_SED = AGN + f_i*3.5E4

###############
### Figures ###
###############
font_size=15.0
label_size=10.
lw = 3.
fig,(ax1) = plt.subplots(1,1,figsize=(20,10),dpi=120)


#Observations
ax1.errorbar(l_Kis,f_Kis,yerr=ferr_Kis,fmt='o',color='red',markersize=10.,label='Kishimoto+09')
ax1.errorbar(l_MMT,f_MMT,yerr=ferr_MMT,fmt='s',color='black',markersize=10.,label='MMT-Pol: This work')
ax1.errorbar(l_Ima,f_Ima,yerr=ferr_Ima,fmt='o',color='grey',markersize=5.,label='Imanishi+10')
ax1.errorbar(l_CCSpec,f_CCSpec,yerr=ferr_CCSpec,fmt='o',color='blue',markersize=5.,label='Alonso-Herrero+16')
ax1.errorbar(l_CCIma,f_CCIma,yerr=ferr_CCIma,fmt='s',color='green',markersize=10.,label='CC-Pol: This work')

#Templates
ax1.plot(templates[3408]['wavesim'],templates[3408]['fluxsim']*1.1e-4,'o',color='black')

#AGN
ax1.plot(l_i,AGN,color='red',linestyle='--',lw=3)

#Clumpy Model
ax1.plot(l_i,f_i*3.5E4,color='green',lw=3)

#TOTAL
ax1.plot(l_i,TOTAL_SED,color='black',lw=5)

ax1.set_xlabel('Wavelength ($\mu$m)',fontsize=font_size)
ax1.set_ylabel('Flux (mJy)',fontsize=font_size)
ax1.set_ylim([0.1,1E5])
ax1.set_xlim([1,13])
ax1.xaxis.set_tick_params(labelsize=font_size)
ax1.yaxis.set_tick_params(labelsize=font_size)
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.legend(loc='best')
```
